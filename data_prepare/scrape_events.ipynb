```python
import requests
from bs4 import BeautifulSoup
import csv
import time

BASE_URL = "https://karnet.krakowculture.pl/wydarzenia?Item_page={}"
OUTPUT_FILE = "wydarzenia.csv"

headers = {
    "User-Agent": "Mozilla/5.0"
}

def get_event_data(soup):
    events = []
    event_divs = soup.find_all("div", class_="event-controls")

    for control in event_divs:
        # Przejście do rodzica, żeby zebrać całość
        parent = control.find_parent("div")

        title_tag = parent.find("h3", class_="event-title")
        type_tag = parent.find("span", class_="event-type")
        location_tag = parent.find("p", class_="event-location")
        date_tag = control.find("a", class_="event-date")

        title = title_tag.get_text(strip=True) if title_tag else ""
        event_type = type_tag.get_text(strip=True) if type_tag else ""
        location = location_tag.get_text(strip=True) if location_tag else ""
        date = date_tag.get_text(strip=True) if date_tag else ""

        events.append([title, event_type, location, date])

    return events

# Zbieranie danych
all_events = []

for page in range(1, 52):
    print(f"Pobieram stronę {page}...")
    url = BASE_URL.format(page)
    response = requests.get(url, headers=headers)
    
    if response.status_code != 200:
        print(f"Błąd pobierania strony {page}: {response.status_code}")
        continue

    soup = BeautifulSoup(response.text, "html.parser")
    page_events = get_event_data(soup)
    all_events.extend(page_events)
    time.sleep(1)  # pauza by nie przeciążyć serwera

# Zapis do CSV
with open(OUTPUT_FILE, mode='w', encoding='utf-8', newline='') as file:
    writer = csv.writer(file, delimiter=';')
    writer.writerow(["Tytuł", "Typ", "Lokalizacja", "Data"])
    writer.writerows(all_events)

print(f"Zapisano {len(all_events)} wydarzeń do pliku {OUTPUT_FILE}")

```
