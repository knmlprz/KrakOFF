```python
import requests
from bs4 import BeautifulSoup
import os
import time
```

```python
url = 'https://www.bip.krakow.pl/?mmi=234'

# Konfiguracja sesji (opcjonalnie, np. dla nagłówków)
session = requests.Session()
session.headers.update({
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
})

# Wysyłanie żądania HTTP
response = requests.get(url)
response.raise_for_status()  # Sprawdzenie, czy żądanie się powiodło

# Parsowanie zawartości strony
soup = BeautifulSoup(response.text, 'html.parser')

# Tworzenie folderu na pliki PDF, jeśli nie istnieje
if not os.path.exists('pdf_files'):
    os.makedirs('pdf_files')
```

```python
# Wyszukiwanie wszystkich linków do plików PDF
for link in soup.find_all('a', href=True):
    href = link['href']
    if href.endswith('.pdf') or '[pdf' in link.text.lower():
        # Pobieranie pełnego URL, jeśli link jest względny
        if not href.startswith('http'):
            href = f"https://www.bip.krakow.pl{href}"  # Zastąp 'domena.pl' odpowiednią domeną
        # Pobieranie nazwy pliku z linku lub tekstu
        filename = os.path.join('pdf_files', href.split('/')[-1] or f"file_{link.text.strip()}.pdf")
        # Pobieranie pliku PDF
        pdf_response = requests.get(href)
        pdf_response.raise_for_status()
        # Zapisywanie pliku
        with open(filename, 'wb') as f:
            f.write(pdf_response.content)
        print(f"Pobrano: {filename}")

print("Zakończono pobieranie plików PDF.")
```
